```python
colunas_usadas = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
```

```python
colunas_usadas
```

```python
base = pd.read_csv('house-prices.csv', usecols = colunas_usadas) # filtra as colunas que nao sao necessarias
```

```python
base.head()
```

```python
#Escalonar todos os elementos (exceto o preço) da tabela para normaliza-los entre 0 e 1
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scaler_x.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']])
```

```python
base.head()
```

```python
#Agora escalonamos o preço
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
base.head()
```

```python
# Vamos criar uma variável que armazena todos os elementos que implicam no preço e outra que contém estes preços
# Para isso, fazemos x receber a base removendo a coluna de preços com o .drop (axis = 1 serve para remover toda a coluna)
x = base.drop('price', axis = 1)
y = base.price
```

```python
x.head()
```

```python
y.head()
```

```python
type(y) # Quando a variável contém apenas uma coluna, ele é do tipo Series
```

```python
type(x) #Caso contrátrio, é um dataframe
```

```python
previsores_colunas = colunas_usadas[1:17]
```

```python
previsores_colunas
```

```python
import tensorflow.compat.v1 as tf
tf.compat.v1.disable_eager_execution()
```

```python
colunas = [tf.feature_column.numeric_column(key = c) for c in previsores_colunas]
```

```python
print(colunas[10])
```

```python
#Dividir a data base para treinamento e teste
from sklearn.model_selection import train_test_split
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x, y, test_size = 0.3) #30% da database sera para o teste
```

```python
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(x = x_treinamento, y = y_treinamento, 
                                                         batch_size = 32, num_epochs = None, shuffle = True)
```

```python
funcao_teste = tf.estimator.inputs.pandas_input_fn(x = x_teste, y = y_teste,  #serve apenas para gerar as metricas de teste
                                                    batch_size = 32, num_epochs = 10000, shuffle = False)
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=colunas)
```

```python
regressor.train(input_fn=funcao_treinamento, steps = 10000)
```

```python
metricas_treinamento = regressor.evaluate(input_fn=funcao_treinamento, steps = 10000)
```

```python
metricas_teste = regressor.evaluate(input_fn=funcao_teste, steps = 10000)
```

```python
metricas_treinamento
```

```python
metricas_teste 
```

```python
funcao_previsao = tf.estimator.inputs.pandas_input_fn(x = x_teste, shuffle = False) #Funcao para prever os preços
```

```python
previsoes = regressor.predict(input_fn=funcao_previsao)
```

```python
list(previsoes) #precos previstos, porem estão escalonados e em formato de dicionario
```

```python
valores_previsoes = [] #vamos transformar em uma array
for p in regressor.predict(input_fn=funcao_previsao):
    valores_previsoes.append(p['predictions'])
```

```python
valores_previsoes
```
