```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution() 
```

```python
writer = tf.summary.create_file_writer('output_v2')
```

```python
@tf.function
def f(x, y):
  return x ** 2 + y
x = tf.constant([2, 3])
y = tf.constant([3, -2])
f(x, y)
```

```python
a = tf.constant(15)
b = tf.constant(5)
prod = tf.multiply(a, b)
soma = tf.add(a, b)
res = tf.divide(prod, soma)
```

```python
session = tf.compat.v1.Session()
```

```python
out = session.run(res)
print(out)
```
